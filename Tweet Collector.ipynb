## Read in Some #SocialDistancing Tweets from the Seattle Area

```python
# This just runs - the only way I have found to stop it is stopping the kernel 
tt = [] # tweet text
tl = [] # tweet location
tc = [] # tweet coordinates
rtt = [] # retweet text

# Locations are set around the greater Seattle Metro Area, looking for #socialdistancing tweets
for tweet in t.filter(track='#reopen'):
    tt.append(tweet['text'])
    tl.append(tweet['user']['location'])
    tc.append(tweet['coordinates'])
    try:
        rtt.append(tweet['retweeted_status']['full_text'])
    except:
        rtt.append('N/A')
```

```python
# Turn the lists of tweets into a dataframe
tweetdf = pd.DataFrame({'Tweet Text':tt,'Location':tl, 'Tweet Coordinates':tc, 'Retweeted Text':rtt})
```
